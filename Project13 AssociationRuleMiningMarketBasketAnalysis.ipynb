```python
# Importing necessary libraries

import pandas as pd
import numpy as np
```

```python
# Reading the store data 
df = pd.read_csv("./store_data.csv",delimiter=',')
df.head(30)
```

```python
from mlxtend.preprocessing import TransactionEncoder

te = TransactionEncoder()
dataset = te.fit(df).transform(df)
dataset = dataset.astype('int')
print(dataset)
```

```python
dataset_df = pd.DataFrame(dataset)
dataset_df
```

```python
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
## return the items and itemsets with at least 0.001% support
from mlxtend.frequent_patterns import apriori
apriori(dataset_df, min_support = 0.0001)
```

```python
frequent_itemsets = apriori(dataset_df, min_support=0.001,use_colnames = True)
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
frequent_itemsets
```

```python
frequent_itemsets[ (frequent_itemsets['length'] == 2) &
                   (frequent_itemsets['support'] >= 0.001) ]
```

```python
rules = association_rules(frequent_itemsets,metric = "lift",min_threshold=1)
rules.head()
```

```python
rules = association_rules(frequent_itemsets, metric='confidence',min_threshold=0.4)
rules.head(10)
```

```python
rules[ (rules['lift'] > 1) & (rules['confidence'] > 0.5)]
```
